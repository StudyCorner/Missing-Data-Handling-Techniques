# **Necessary Imports**

```python
# necessary imports

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px

import warnings
warnings.filterwarnings('ignore')

plt.style.use('fivethirtyeight')
%matplotlib inline
pd.set_option('display.max_columns', 26)
```

# **loading data**

```python
# loading data

ckd= pd.read_csv('kidney_disease.csv')
ckd.head()
```

```python
ckd.shape
```

# **Dropping Id Column**

```python
# dropping id column
ckd.drop('id', axis = 1, inplace = True)
```

# **Rename Column Names**

```python
# rename column names to make it more user-friendly

ckd.columns = ['age', 'blood_pressure', 'specific_gravity', 'albumin', 'sugar', 'red_blood_cells', 'pus_cell',
              'pus_cell_clumps', 'bacteria', 'blood_glucose_random', 'blood_urea', 'serum_creatinine', 'sodium',
              'potassium', 'haemoglobin', 'packed_cell_volume', 'white_blood_cell_count', 'red_blood_cell_count',
              'hypertension', 'diabetes_mellitus', 'coronary_artery_disease', 'appetite', 'peda_edema',
              'aanemia', 'class']
```

```python
ckd.head()
```

```python
ckd.describe()
```

```python
ckd.info()
```

# As we can see that 'packed_cell_volume', 'white_blood_cell_count' and 'red_blood_cell_count' are object type. We need to change them to numerical dtype

# **Converting Necessary Columns to Numerical Type**

```python
# converting necessary columns to numerical type

ckd['packed_cell_volume'] = pd.to_numeric(ckd['packed_cell_volume'], errors='coerce')
ckd['white_blood_cell_count'] = pd.to_numeric(ckd['white_blood_cell_count'], errors='coerce')
ckd['red_blood_cell_count'] = pd.to_numeric(ckd['red_blood_cell_count'], errors='coerce')
```

```python
ckd.info()
```

# **Extracting categorical and numerical columns**

```python
# Extracting categorical and numerical columns

cat_cols = [col for col in ckd.columns if ckd[col].dtype == 'object']
num_cols = [col for col in ckd.columns if ckd[col].dtype != 'object']
```

# **looking at unique values in categorical columns**

```python
# looking at unique values in categorical columns

for col in cat_cols:
    print(f"{col} has {ckd[col].unique()} values\n")
```

# There is some ambugity present in the columns we have to remove that.

```python
# replace incorrect values

ckd['diabetes_mellitus'].replace(to_replace = {'\tno':'no','\tyes':'yes',' yes':'yes'},inplace=True)

ckd['coronary_artery_disease'] = ckd['coronary_artery_disease'].replace(to_replace = '\tno', value='no')

ckd['class'] = ckd['class'].replace(to_replace = {'ckd\t': 'ckd', 'notckd': 'not ckd'})
```

```python
ckd['class'] = ckd['class'].map({'ckd': 0, 'not ckd': 1})
ckd['class'] = pd.to_numeric(ckd['class'], errors='coerce')
```

```python
cols = ['diabetes_mellitus', 'coronary_artery_disease', 'class']

for col in cols:
    print(f"{col} has {ckd[col].unique()} values\n")
```

# KNN Imputer Object Creation

```python
# create an object for KNNImputer
imputer = KNNImputer(n_neighbors=2)

# fit the imputer on the data
imputer.fit(ckd)
After_imputation = imputer.fit_transform(ckd)
```

```python
ckd_imputed = pd.DataFrame(After_imputation, columns = ckd.columns)
ckd_imputed
```

# **Feature Engineering**

```python
ind_col_ckd_imputed = [col for col in ckd_imputed.columns if col != 'class']
dep_col_ckd_imputed = 'class'

X = ckd_imputed[ind_col_ckd_imputed]
y = ckd_imputed[dep_col_ckd_imputed]
```

# **Decision Tree Classifier**

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report
```

**Building Model with Training Data**

```python
dtc = DecisionTreeClassifier()
dtc.fit(X, y)
```

```python
print(f"Training Accuracy of Decision Tree Classifier is {accuracy_score(y, dtc.predict(X))}")
```

```python
print(f"Confusion Matrix for Testing Data:- \n{accuracy_score(y, dtc.predict(X))}\n")
```

```python
print(f"Classification Report for Testing Data:- \n {classification_report(y, dtc.predict(X))}")
```

# **GridSearchCV**

```python
# hyper parameter tuning of decision tree

from sklearn.model_selection import GridSearchCV
grid_param = {
    'criterion' : ['gini', 'entropy'],
    'max_depth' : [3, 5, 7, 10],
    'splitter' : ['best', 'random'],
    'min_samples_leaf' : [1, 2, 3, 5, 7],
    'min_samples_split' : [1, 2, 3, 5, 7],
    'max_features' : ['auto', 'sqrt', 'log2']
}

grid_search_dtc = GridSearchCV(dtc, grid_param, cv = 5, n_jobs = -1, verbose = 1)
grid_search_dtc.fit(X, y)
```

```python
# best parameters and best score

print(grid_search_dtc.best_params_)
print(grid_search_dtc.best_score_)
```

```python
# best estimator

dtc = grid_search_dtc.best_estimator_

# accuracy score, confusion matrix and classification report of decision tree

dtc_acc = accuracy_score(y, dtc.predict(X))
```

```python
print(f" Accuracy of Decision Tree Classifier is {dtc_acc}")
```

```python
print(f"Confusion Matrix :- \n{confusion_matrix(y, dtc.predict(X))}\n")
```

```python
print(f"Classification Report :- \n {classification_report(y, dtc.predict(X))}")
```
